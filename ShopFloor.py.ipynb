```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

DATA_DIR = './'
PATHS = '{0}/epaths.csv'.format(DATA_DIR)
```

```python
paths = pd.read_csv(PATHS)
paths.head(28)
```
